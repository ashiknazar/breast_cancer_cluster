```python
import pandas as pd
```

```python
data=pd.read_csv('3gene_filled.csv')
```

```python
cols=list(data.columns.values)
for i in cols:
    if(data[i].isna().sum()!=0):
        print(i,data[i].isna().sum())
```

```python
import pandas as pd
from sklearn.experimental import enable_iterative_imputer  # noqa
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import LabelEncoder


df = data

# Step 1: Convert categorical 'cellularity' to numeric
label_encoder = LabelEncoder()
df['cellularity_encoded'] = label_encoder.fit_transform(df['cellularity'].astype(str))

# Step 2: Prepare the data for imputation
# Include 'cellularity_encoded' in the imputation process
imputer = IterativeImputer(random_state=0)
df[['neoplasm_histologic_grade', 
    'lymph_nodes_examined_positive', 
    'nottingham_prognostic_index', 
    'cellularity_encoded']] = imputer.fit_transform(
    df[['neoplasm_histologic_grade', 
        'lymph_nodes_examined_positive', 
        'nottingham_prognostic_index',  
        'cellularity_encoded']]
)

# Step 3: Convert numeric values back to categorical
df['cellularity'] = label_encoder.inverse_transform(df['cellularity_encoded'].round().astype(int))

# Drop the encoded column if no longer needed
df.drop(columns=['cellularity_encoded'], inplace=True)

# Display the DataFrame
df
```

```python
df.to_csv('third.csv',index=False)
```

```python
import pandas as pd
from sklearn.experimental import enable_iterative_imputer  # noqa
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import LabelEncoder




label_encoder = LabelEncoder()
df['primary_tumor_laterality_encoded'] = label_encoder.fit_transform(df['primary_tumor_laterality'].astype(str))

# Step 2: Prepare the data for imputation
features_to_impute = ['age_at_diagnosis', 
                      'nottingham_prognostic_index', 
                      'lymph_nodes_examined_positive', 
                      'neoplasm_histologic_grade', 
                      'primary_tumor_laterality_encoded']

imputer = IterativeImputer(random_state=0)

# Fit and transform the selected features
df[features_to_impute] = imputer.fit_transform(df[features_to_impute])

# Step 3: Convert numeric values back to categorical
df['primary_tumor_laterality'] = label_encoder.inverse_transform(df['primary_tumor_laterality_encoded'].round().astype(int))

# Cleanup: Drop the encoded column
df.drop(columns=['primary_tumor_laterality_encoded'], inplace=True)

# Display the DataFrame
df
```
